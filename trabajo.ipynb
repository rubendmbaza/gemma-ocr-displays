```python
from transformers import AutoProcessor, AutoModelForVision2Seq
from PIL import Image

# Cargar modelo y procesador (versin 4B multimodal)
processor = AutoProcessor.from_pretrained("google/gemma-4b-multimodal")
model = AutoModelForVision2Seq.from_pretrained("google/gemma-4b-multimodal", device_map="auto")
```

```python
def read_display_gemma(image_path):
    # Cargar y preprocesar imagen
    image_path = 'train/images/00b527335f956f6e4b1ba87ae0fa96d6b2b82314_jpg.rf.586ae70ec06b591e56ff8a385c23c3fb.jpg'
    image = Image.open(image_path).convert("RGB")

    # Crear prompt especfico
    prompt = "You are an industrial display reader. What number is shown in this 7-segment display? Answer only with the numerical value:"

    # Procesar entrada multimodal
    inputs = processor(text=prompt, images=image, return_tensors="pt").to("cuda")

    # Generar respuesta
    outputs = model.generate(**inputs, max_new_tokens=10)
    return processor.decode(outputs[0], skip_special_tokens=True).strip()
```
